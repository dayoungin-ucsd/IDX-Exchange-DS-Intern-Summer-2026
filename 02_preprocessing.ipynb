# **Handle missing values**

```python
missing = pd.DataFrame({
    "Missing Count": df_sf[financial_features].isna().sum(),
    "Missing %": (
        df_sf[financial_features]
        .isna()
        .mean()*100
    ).round(2)
}).sort_values(by="Missing %", ascending=False)

missing
```

```python
df_sf = df_sf.drop(columns=['TaxAnnualAmount', 'TaxYear', 'LotSizeDimensions', 'AssociationFeeFrequency'])

df_sf
```

## **Notes**

**TaxAnnualAmount** and **TaxYear** were dropped because both variables contained 100% missing values, providing no usable information for preprocessing or modeling. **LotSizeDimensions** and **AssociationFeeFrequency** were also dropped because they also had extremely high missing rates (93.67% and 74.55%, respectively).

```python
fill_cols = ['AssociationFee', 'LotSizeSquareFeet', 'LotSizeAcres', 'LotSizeArea', 'OriginalListPrice']

for col in fill_cols:
  df_sf[col] = df_sf[col].fillna(df_sf[col].median())
```

## **Notes**

Missing values in **AssociationFee, LotSizeSquareFeet, LotSizeAcres, LotSizeArea, and OriginalListPrice** were imputed using the median because these variables are numerical and have outliers. **ListPrice** and **ClosePrice** were kept without modification because they didn't contain any missing values.
